```r
for (sample in samples) {
    G_all = list()
    SBM_all = list()
    col_all = list()
    l_all = list()

    out_dir = sprintf('%s/%s/', ind_dir, sample)
    fns = list.files(out_dir)
    mtx_file = fns[grepl('knn.*mtx', fns)]
    mat = readMM(paste0(out_dir, mtx_file))
    mat = (mat+t(mat))/2

    obs_fn = fns[grepl('obs',fns)]
    obs_df = read.table(paste0(out_dir, obs_fn), header = T, row.names = 1, sep = '\t')

    BL = obs_df %>% dplyr::select(Basal.Correlation) %>% as.matrix

    G = graph_from_adjacency_matrix(mat, mode = 'undirected', weighted =T)
    clusters = cluster_spinglass(G, spins=2)
    print(table(paste(clusters$membership, obs_df$Basal.vs.Luminal.by.Correlation)))

    mat2 = mat !=0

    Out =FitAttribute(mat2, BL, 0, 2, as.numeric(factor(obs_df$Basal.vs.Luminal.by.Correlation)))

    results = list()
    results$SBM = Out
    results$G = G
    saveRDS(results, paste0(out_dir, 'metacells.SBM.RDS'))

    SBM_cluster = Out$Comm
    names(SBM_cluster) = rownames(obs_df)

    write.table(data.frame(SBM_cluster), file = paste0(out_dir, 'SBM_cluster.txt'), sep = ',', quote = F, row.names = T)
}
```

```r
RunGSEA <- function(deg_file, gset0) {
    set.seed(10)
  deg_df = read.table(deg_file, header=T, sep =',', row.names = 'primerid')
  min_fdr = sort(unique(deg_df$fdr))[2]
  deg_df$fdr[deg_df$fdr==0] = min_fdr
  rnk = -log2(deg_df$fdr) * (2*(deg_df$coef > 0) - 1)

  names(rnk) = rownames(deg_df)
  score = rnk[!is.na(rnk)] 
  
  fgseaRes <- fgsea(pathways = gset0, 
              stats = score,
              minSize=10,
              maxSize=500)#,
              #nperm=10000)
  
  fgseaRes$leadingEdge = sapply(fgseaRes$leadingEdge, function(x) str_c(x,collapse=','))
  return(fgseaRes)
  
}

```

```r
deg_df[,'Bonferroni.Padj'] = pmin(1,deg_df[,'Pr..Chisq.']*nrow(deg_df))
```

```r
ind = deg_df$Bonferroni.Padj < 0.05 & abs(deg_df$coef) > log2(1.5)
```
