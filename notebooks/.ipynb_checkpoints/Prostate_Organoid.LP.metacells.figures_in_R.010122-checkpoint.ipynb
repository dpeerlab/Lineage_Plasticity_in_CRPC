````r


G_all = list()
SBM_all = list()
col_all = list()
l_all = list()
max_mc = 0
for (sample in samples) {
  out_dir = sprintf('%s/%s/', ind_dir, sample)
  fns = list.files(out_dir)
  mtx_file = fns[grepl('knn.*mtx', fns)]
  mat = readMM(paste0(out_dir, mtx_file))
  mat = (mat+t(mat))/2
  
  obs_fn = fns[grepl('obs',fns)]
  obs_df = read.table(paste0(out_dir, obs_fn), header = T, row.names = 1, sep = '\t')

  BL = obs_df %>% dplyr::select(Basal.Correlation) %>% as.matrix
    
  G = graph_from_adjacency_matrix(mat, mode = 'undirected', weighted =T)
  clusters = cluster_spinglass(G, spins=2)
  print(sample)
  print(table(paste(clusters$membership, obs_df$Basal.vs.Luminal.by.Correlation)))
  
  mat2 = mat !=0
  
  Out =FitAttribute(mat2, BL, 0, 2, as.numeric(factor(obs_df$Basal.vs.Luminal.by.Correlation)))
  
  col_bl = c('blue','red')[factor(obs_df$Basal.vs.Luminal.by.Correlation)]
  col_clust = c('#C5E5E7','#ECD89A')[factor(Out$Comm)]
  rbPal <- colorRampPalette(c('blue','red'))
  col_bl_corr <- rbPal(10)[as.numeric(cut(obs_df$Basal.Correlation,breaks = 10))]
  
  l = as.matrix(read.table(paste0(out_dir, sprintf('draw_graph_fa.metacells.%s.010122.txt', sample)), sep = '\t', row.names = 1, header = T))
  
  G_all[[sample]] = G
  SBM_all[[sample]] = Out
  l_all[[sample]] = l

  mc_id = as.numeric(rownames(l_all[[sample]])) + max_mc
  max_mc = max(mc_id) + 1
    
  print(unique(obs_df[as.character(mc_id), 'batch']))
  col_all[[sample]] <- col_bl_corr_all[as.character(mc_id)]
  
}


```





```{r}



G_all = list()
SBM_all = list()
col_all = list()
l_all = list()
max_mc = 0
for (sample in samples) {
  # out_dir = sprintf('/Users/chanj3/Documents_NotToICloud/MSKCC/Research/SCPC_transformation.metacells.120120/out.cell_line.individual.120120/%s/', sample)
  # results = readRDS(paste0(out_dir, 'metacells.SBM.RDS'))
  results = results_all[[sample]]
  
  G_all[[sample]] = results$G
  SBM_all[[sample]] = results$SBM
  
  out_dir = sprintf('%s/%s/', ind_dir, sample)
  l = as.matrix(read.table(paste0(out_dir, sprintf('draw_graph_fa.metacells.%s.010122.txt', sample)), sep = '\t', row.names = 1, header = T))

  l_all[[sample]] = l

  mc_id = as.numeric(rownames(l_all[[sample]])) + max_mc
  max_mc = max(mc_id) + 1
    
  print(unique(obs_df[as.character(mc_id), 'batch']))
  col_all[[sample]] <- col_bl_corr[as.character(mc_id)]
  
}

```



```{r}


par(mfrow= c(3,2), mar = c(0,0,1,0)) 
for (sample in names(G_all)) {
  plot(G_all[[sample]], vertex.color = col_all[[sample]], vertex.size = 8, vertex.label = NA, layout = l_all[[sample]], main = gsub('CRE','Week',sample),
       mark.groups=list(which(SBM_all[[sample]]$Comm==1), which(SBM_all[[sample]]$Comm==2)), mark.col = rev(rainbow(2, alpha = 0.1)), mark.border= NA) #mark.col=c("#ECD89A", "#C5E5E7"), 
}




```



```{r}

# Function to plot color bar
color.bar <- function(lut, min, max=-min, nticks=5, ticks=seq(min, max, len=nticks), title='') {
    scale = (length(lut)-1)/(max-min)

    #dev.new(width=1.75, height=5)
    plot(c(0,1), c(min,max), type='n', bty='n', xaxt='n', xlab='', yaxt='n', ylab='', main=title)
    axis(2, ticks, las=1, cex.axis = 1.5)
    for (i in 1:(length(lut)-1)) {
     y = (i-1)/scale + min
     rect(0,y,10,y+1/scale, col=lut[i], border=NA)
    }
}

```


```{r}

min = min(obs_df$Basal.Correlation)
max = max(obs_df$Basal.Correlation)
nticks = 11

color.bar(colorRampPalette(c('blue','yellow','red'))(100), min = -0.2)

```



### FIGURE 2H
```{r}

comb_dir = '/Users/chanj3/Documents_NotToICloud/MSKCC/Research/Prostate_Organoid.LP.publication.010122/out.metacells.combined.010122/'

plot_df = read.table(paste0(comb_dir, 'diffusion_distance.BL.sampled.txt'), sep = '\t', header =T, row.names = 1)

plot_df$Sample = as.factor(plot_df$Sample)

plot_df %>% ggplot(aes(Sample,Diffusion_Distance,fill=Sample)) + geom_boxplot(outlier.shape = NA, lwd = .1) +
  theme_bw() + ggtitle('Basal-luminal plasticity') + ylab('Log Mean\nDiffusion Distance') + theme(text = element_text(size=15), axis.text.x = element_text(angle=30, vjust=1, hjust=1),
                                                                                                         panel.grid.major = element_blank(),
                                                                                                         panel.grid.minor = element_blank(),
                                                                                                         legend.position = "none") 

```


```{r}

for (i in 1:5) {
  tp1 = levels(plot_df$Sample)[i]
  tp2 = levels(plot_df$Sample)[i+1]
  
  dd1 = plot_df %>% dplyr::filter(Sample == tp1) %>% dplyr::select(Diffusion_Distance)
  dd2 = plot_df %>% dplyr::filter(Sample == tp2) %>% dplyr::select(Diffusion_Distance)
  
  #res = wilcox.test(dd1$Diffusion_Distance, dd2$Diffusion_Distance, alternative = "two.sided")
  
  #res = t.test(dd1$Diffusion_Distance, dd2$Diffusion_Distance, alternative = "two.sided", var.equal = FALSE)
  res = chisq.test(dd1$Diffusion_Distance, dd2$Diffusion_Distance) 
  
  print(res$p.value * 5)
}

```

```{r}

tp1 = levels(plot_df$Sample)[3]
tp2 = levels(plot_df$Sample)[5]
  
dd1 = plot_df %>% dplyr::filter(Sample == tp1) %>% dplyr::select(Diffusion_Distance)
dd2 = plot_df %>% dplyr::filter(Sample == tp2) %>% dplyr::select(Diffusion_Distance)

res = wilcox.test(dd1$Diffusion_Distance, dd2$Diffusion_Distance, alternative = "two.sided")

res

```


### FIGURE 3A, S5A
```{r}

library(SingleCellExperiment)
library(zellkonverter)

```



```{r}

RunGSEA <- function(deg_file, gset0) {

  deg_df = read.table(deg_file, header=T, sep =',', row.names = 'primerid')
  min_fdr = sort(unique(deg_df$fdr))[2]
  deg_df$fdr[deg_df$fdr==0] = min_fdr
  rnk = -log2(deg_df$fdr) * (2*(deg_df$coef > 0) - 1)
  # rnk = -log2(deg_df$fdr) * deg_df$coef
  names(rnk) = rownames(deg_df)
  score = rnk[!is.na(rnk)] 
  
  fgseaRes <- fgsea(pathways = gset0, 
              stats = score,
              minSize=10,
              maxSize=500)#,
              #nperm=10000)
  
  fgseaRes$leadingEdge = sapply(fgseaRes$leadingEdge, function(x) str_c(x,collapse=','))
  return(fgseaRes)
  
}


```

```{r}

gsea_dir = '/Users/chanj3/Documents_NotToICloud/MSKCC/Research/Prostate_Organoid.LP.publication.010122/out.metacells.combined.010122/GSEA.MAST.010122/'
deg_dir2 = paste0(comb_dir, 'MAST.MUT_WT/')
deg_fn = 'MUT_WT.MUT.csv'

fgseaRes = RunGSEA(paste0(deg_dir2, deg_fn), gset)


```

```{r}

gsea_fn = 'GSEA.MUT_WT.csv'

fgseaRes = read.table(paste0(gsea_dir, gsea_fn), sep = '\t', header = T)  


```


```{r}

pval_cutoff = 0.1

ind1 = fgseaRes$padj < pval_cutoff
ind2 = abs(fgseaRes$NES) > 1
ind1[is.na(ind1)] = FALSE
ind2[is.na(ind2)] = FALSE

plot_df = fgseaRes[ind1 & ind2,]


```


```{r}

pathways_select = readLines(paste0(gsea_dir, 'pathways.MUT_WT.select.txt'))

```


```{r}

hm_df = fgseaRes %>% dplyr::filter((padj < 0.05 & NES < -1) | pathway %in% pathways_select | grepl('Org', pathway)) %>% arrange(NES)

```


```{r}

sce = zellkonverter::readH5AD(paste0(comb_dir, 'adata.metacells.combined.010122.h5ad'))

data_df = t(assays(sce)$X)
data_df = scale(data_df)

obs_df = read.table(paste0(comb_dir, 'obs.metacells.combined.010122.txt'), header=T, sep='\t', row.names = 1)

hm_mat = matrix(0, nrow(obs_df), nrow(hm_df))
colnames(hm_mat) = hm_df$pathway
rownames(hm_mat) = rownames(obs_df)
for (i in hm_df$pathway) {
  LE = str_split(hm_df[hm_df$pathway == i, 'leadingEdge'],',')[[1]]
  mat = rowMeans(data_df[,colnames(data_df) %in% LE])
  hm_mat[,i] = mat
}


```


```{r}

hm_mat = matrix(0, 6, nrow(hm_df))
colnames(hm_mat) = hm_df$pathway
rownames(hm_mat) = sort(unique(obs_df$batch))


for (i in hm_df$pathway) {
  LE = str_split(hm_df[hm_df$pathway == i, 'leadingEdge'],',')[[1]]
  mat = aggregate(data.frame(data_df[,colnames(data_df) %in% LE]), list(obs_df$batch), mean)
  rownames(mat) = mat$Group.1
  mat = mat[,-1]
  hm_mat[,i] = rowMeans(mat)
}

colnames(hm_mat)

hm_mat = hm_mat[,!grepl( 'REACTOME_APC_C_',colnames(hm_mat))]

```


```{r}

obs_df$batch_bl = paste0(gsub('CD24', 'Luminal', gsub('CD49f', 'Basal', obs_df$Basal.vs.Luminal.by.Correlation)), ': ', obs_df$batch)

hm_mat = matrix(0, 12, nrow(hm_df))
colnames(hm_mat) = hm_df$pathway
rownames(hm_mat) = sort(unique(obs_df$batch_bl))
for (i in hm_df$pathway) {
  LE = str_split(hm_df[hm_df$pathway == i, 'leadingEdge'],',')[[1]]
  mat = stats::aggregate(data.frame(data_df[,LE[LE %in% colnames(data_df)]]), list(obs_df$batch_bl), mean)
  rownames(mat) = mat$Group.1
  mat = mat[,-1]
  hm_mat[,i] = rowMeans(mat)
}

hm_mat = hm_mat[,colnames(hm_mat) != 'PID_MYC_ACTIV_PATHWAY']
hm_mat = hm_mat[,colnames(hm_mat) != 'HALLMARK_WNT_BETA_CATENIN_SIGNALING']
hm_mat = hm_mat[,colnames(hm_mat) !="REACTOME_APC_C_CDC20_MEDIATED_DEGRADATION_OF_MITOTIC_PROTEINS" ]
hm_mat = hm_mat[,colnames(hm_mat) !="REACTOME_INTERFERON_ALPHA_BETA_SIGNALING" ]
hm_mat = hm_mat[,colnames(hm_mat) !="HALLMARK_KRAS_SIGNALING_DN" ]
hm_mat = hm_mat[,colnames(hm_mat) !="PID_SMAD2_3NUCLEAR_PATHWAY" ]
hm_mat = hm_mat[,colnames(hm_mat) !="HALLMARK_HYPOXIA" ]

```


```{r}

gene_overlap = matrix(0, nrow(hm_df), nrow(hm_df))
colnames(gene_overlap) = hm_df$pathway
rownames(gene_overlap) = hm_df$pathway

rownames(hm_df) = hm_df$pathway

for (p1 in hm_df$pathway) {
  for (p2 in hm_df$pathway) {
    le1 = str_split(hm_df[p1, 'leadingEdge'],',')[[1]]
    le2 = str_split(hm_df[p2, 'leadingEdge'],',')[[1]]
    gene_overlap[p1, p2] = length(intersect(le1, le2)) / length(union(le1, le2))
  }
}


```

```{r}

pathway_categories = read.table(sprintf('%s/pathway_categories.txt',gsea_dir), sep ='\t', header=T, row.names=1)
pathway_categories$Category = gsub('Other ','Other\n',pathway_categories$Category)
pathway_categories$Category = gsub('/','/\n',pathway_categories$Category)
pathway_categories = pathway_categories[colnames(hm_mat),]
names(pathway_categories) = colnames(hm_mat)

names(pathway_categories) = gsub('REACTOME_ANTIGEN_PROCESSING_UBIQUITINATION_PROTEASOME_DEGRADATION','REACTOME_UBIQUITINATION_PROTEASOME_DEGRADATION',names(pathway_categories))
names(pathway_categories) = gsub('REACTOME_APC_C_CDC20_MEDIATED_DEGRADATION_OF_MITOTIC_PROTEINS','REACTOME_APC_C_CDC20_DEGRADING_OF_MITOTIC_PROTEINS',names(pathway_categories))
names(pathway_categories) = gsub('REACTOME_P53_INDEPENDENT_G1_S_DNA_DAMAGE_CHECKPOINT','REACTOME_P53_INDEPT_G1_S_DNA_DAMAGE_CHECKPOINT',names(pathway_categories))

pathway_categories = factor(pathway_categories)

colnames(hm_mat) = gsub('REACTOME_ANTIGEN_PROCESSING_UBIQUITINATION_PROTEASOME_DEGRADATION','REACTOME_UBIQUITINATION_PROTEASOME_DEGRADATION',colnames(hm_mat))
colnames(hm_mat) = gsub('REACTOME_APC_C_CDC20_MEDIATED_DEGRADATION_OF_MITOTIC_PROTEINS','REACTOME_APC_C_CDC20_DEGRADING_OF_MITOTIC_PROTEINS',colnames(hm_mat))
colnames(hm_mat) = gsub('REACTOME_P53_INDEPENDENT_G1_S_DNA_DAMAGE_CHECKPOINT','REACTOME_P53_INDEPT_G1_S_DNA_DAMAGE_CHECKPOINT',colnames(hm_mat))

colnames(gene_overlap) = gsub('REACTOME_ANTIGEN_PROCESSING_UBIQUITINATION_PROTEASOME_DEGRADATION','REACTOME_UBIQUITINATION_PROTEASOME_DEGRADATION', colnames(gene_overlap))
colnames(gene_overlap) = gsub('REACTOME_APC_C_CDC20_MEDIATED_DEGRADATION_OF_MITOTIC_PROTEINS','REACTOME_APC_C_CDC20_DEGRADING_OF_MITOTIC_PROTEINS',colnames(gene_overlap))
colnames(gene_overlap) = gsub('REACTOME_P53_INDEPENDENT_G1_S_DNA_DAMAGE_CHECKPOINT','REACTOME_P53_INDEPT_G1_S_DNA_DAMAGE_CHECKPOINT',colnames(gene_overlap))
rownames(gene_overlap) = colnames(gene_overlap)



```


```{r}

library(pals)

num_col = length(levels(pathway_categories))
col = watlington(num_col)
names(col) = levels(pathway_categories)

```


```{r, fig.width = 5, fig.height=5}

library(dendextend)
library(ComplexHeatmap)
library(circlize)

hm_mat_b = t(hm_mat)[,grepl('Basal',rownames(hm_mat))]
colnames(hm_mat_b) = gsub('Basal: ', '', colnames(hm_mat_b))
hm_mat_l = t(hm_mat)[,grepl('Luminal',rownames(hm_mat))]
colnames(hm_mat_l) = gsub('Luminal: ', '', colnames(hm_mat_b))


samples = sort(unique(obs_df$batch))

dd = as.dist(1-cor(hm_mat_b))
column_dend_b = as.dendrogram(hclust(dd))
column_dend_b = rotate(column_dend_b, order = samples) 

dd = as.dist(1-cor(hm_mat_l))
column_dend_l = as.dendrogram(hclust(dd))
column_dend_l = rotate(column_dend_l, order = samples) 


hm = Heatmap(hm_mat_l, name = "expression", 
    col = colorRamp2(c(-1, 0, 1), c('blue3','white','red3')),
    cluster_rows = T, cluster_columns = column_dend_l, #column_split = gsub(': .*','',rownames(hm_mat)), 
    clustering_distance_rows = "pearson", row_split = pathway_categories,
    clustering_distance_columns = "pearson",
    show_row_names = T, 
    show_column_names = T, column_title = 'Luminal', column_title_side = "bottom", show_column_dend=T, row_title_rot = 0, row_title_gp = gpar(fontsize=13, fontface = "bold"), 
    heatmap_legend_param = list(direction = 'horizontal'))  + 
  Heatmap(hm_mat_b, name = "expression", 
    col = colorRamp2(c(-1, 0, 1), c('blue3','white','red3')),
    cluster_rows = T, cluster_columns = column_dend_b, #column_split = gsub(': .*','',rownames(hm_mat)), 
    clustering_distance_rows = "pearson", row_split = pathway_categories,
    clustering_distance_columns = "pearson",
    show_row_names = T,
    show_column_names = T, column_title = 'Basal', column_title_side = "bottom", show_column_dend=T,
    heatmap_legend_param = list(direction = 'horizontal'))  +
  Heatmap(pathway_categories, name = "Pathways", col = col,
    show_heatmap_legend = F, width = unit(5, "mm"), heatmap_legend_param = list(ncol = 2))

draw(hm, padding = unit(c(2, 2, 2, 70), "mm"), heatmap_legend_side = "bottom")



```




```{r}

out_fn = paste0(gsea_dir, 'GSEA.MUT_WT.filtered.csv')

hm_df2 = read.table(out_fn, sep = '\t', header = T)

p_interest2 =  c("ABBUD_LIF_SIGNALING_1_UP",
                 "FGFR_signature.Acevedo_etal",
                 "HALLMARK_ANDROGEN_RESPONSE",
                 "HALLMARK_APOPTOSIS",
                 "HALLMARK_INTERFERON_GAMMA_RESPONSE",
                 "HALLMARK_KRAS_SIGNALING_UP",
                 "KEGG_JAK_STAT_SIGNALING_PATHWAY",
                 "PID_SMAD2_3NUCLEAR_PATHWAY",
                 "REACTOME_APC_C_CDH1_MEDIATED_DEGRADATION_OF_CDC20_AND_OTHER_APC_C_CDH1_TARGETED_PROTEINS_IN_LATE_MITOSIS_EARLY_G1",
                 "REACTOME_INTERFERON_ALPHA_BETA_SIGNALING",
                 "WNT_UP.V1_DN")                                                                                                

```


```{r}

hm_df3 = hm_df2 %>% dplyr::filter(pathway %in% p_interest2) #%>% dplyr::select(leadingEdge)
rownames(hm_df3) = hm_df3$pathway

hm_mat5 = matrix(0, 6, nrow(hm_df3))
colnames(hm_mat5) = rownames(hm_df3)
rownames(hm_mat5) = sort(unique(obs_df$batch))
for (i in rownames(hm_df3)) {
  LE = str_split(hm_df3[i, 'leadingEdge'],',')[[1]]
  mat = stats::aggregate(data.frame(data_df[,LE[LE %in% colnames(data_df)]]), list(obs_df$batch), mean)
  rownames(mat) = mat$Group.1
  mat = mat[,-1]
  hm_mat5[,i] = rowMeans(mat)
}

colnames(hm_mat5)[colnames(hm_mat5)=="REACTOME_APC_C_CDH1_MEDIATED_DEGRADATION_OF_CDC20_AND_OTHER_APC_C_CDH1_TARGETED_PROTEINS_IN_LATE_MITOSIS_EARLY_G1"] = "REACTOME_APC_C_CDH1_MEDIATED_DEGRADATION"

```



```{r, fig.width=4, fig.height=3}

samples = c("DHT 0", "DHT 2", "DHT 4", "DHT 8", "ENZ 4", "ENZ 8")

hm = Heatmap(t(hm_mat5[samples,]), name = "expression", 
    col = colorRamp2(c(-.3, 0, .5), c('blue3','yellow','red3')),
    row_split = 4, #clusters[colnames(hm_mat5)],
    column_split = c(1,2,3,3,3,3),
    cluster_rows = T, cluster_columns = F,
    clustering_distance_rows = "pearson",
    clustering_distance_columns = "pearson",
    show_row_names = T, column_title = NULL, 
    show_column_names = T, row_title = NULL, show_column_dend=T,
    heatmap_legend_param = list(direction = 'horizontal'))  

hm = draw(hm, heatmap_legend_side = "bottom", padding = unit(c(2, 2, 2, 80), "mm"))


```


```{r}

ord = order(hm_mat5['DHT 2',])

hm = Heatmap(t(hm_mat5[samples,rev(ord)]), name = "expression", 
    col = colorRamp2(c(-.3, 0, .5), c('blue3','yellow','red3')),
    column_split = c(1,2,3,3,3,3),
    cluster_rows = F, cluster_columns = F,
    show_row_names = T, column_title = NULL, 
    show_column_names = T, row_title = NULL, show_column_dend=T,
    heatmap_legend_param = list(direction = 'horizontal'))  

hm = draw(hm, heatmap_legend_side = "bottom", padding = unit(c(2, 2, 2, 80), "mm"))


```



### FIGURE 3C
```{r}

library(stringr)
library(tibble)

gsea_dir = '/Users/chanj3/Documents_NotToICloud/MSKCC/Research/Prostate_Organoid.LP.publication.010122/out.metacells.combined.010122/GSEA.MAST.010122/'
gsea_fn = paste0(gsea_dir, 'GSEA.MUT_WT.csv')

gsea_df = read.table(gsea_fn, sep='\t', header=T, stringsAsFactors = F, row.names = 'pathway')

pval_cutoff = 0.05

ind1 = gsea_df$padj < pval_cutoff
ind2 = abs(gsea_df$NES) > 1
ind1[is.na(ind1)] = FALSE
ind2[is.na(ind2)] = FALSE

gsea_df = gsea_df[ind1 & ind2,]

le_genes1=  str_split(gsea_df['KEGG_JAK_STAT_SIGNALING_PATHWAY', 'leadingEdge'], ',')[[1]]
le_genes2=  str_split(gsea_df['HALLMARK_IL6_JAK_STAT3_SIGNALING', 'leadingEdge'], ',')[[1]]
le_genes = union(le_genes1, le_genes2)

mat3 = scale(mat[,le_genes])

jak_stat_rho_per_gene = data.frame(rho = t(cor(H, mat3))) %>% rownames_to_column('gene') %>% arrange(-rho)

jak_stat_score = rowMeans(mat3)



```


```{r}

mc_dir = '~/Documents_NotToICloud/MSKCC/Research/Prostate_Organoid.LP.publication.010122/out.metacells.combined.010122/'
mat_mc = read.table(paste0(mc_dir, 'norm_df.metacells.combined.010122.csv'), sep ='\t', stringsAsFactors = F, header = T, row.names = 1)

mat_mc2 = apply(mat_mc[,le_genes], 2, scale)

mat_mc3 = mat_mc2  #[,jak_stat_rho_per_gene$gene[1:13]]

jak_stat_score_mc = rowMeans(mat_mc3)

sc_dir = '/Users/chanj3/Documents_NotToICloud/MSKCC/Research/Prostate_Organoid.LP.publication.010122/out.single_cell.combined.010122/'
obs_sc_df = read.table(sprintf('%sobs.single_cell.combined.010122.txt',sc_dir), header=T, stringsAsFactors = F, sep = '\t', row.names = 'index')

plot_df = data.frame(H=H, mc = obs_sc_df$metacells) 
plot_df$mc = factor(plot_df$mc) 
plot_df = aggregate(plot_df$H, list(plot_df$mc), mean) %>% dplyr::rename(Entropy = x, metacell = Group.1) 

obs_mc_df = read.table(paste0(mc_dir, 'obs.metacells.combined.010122.txt'), sep = '\t', header=T)

plot_df$DPT  = obs_mc_df$DPT
plot_df$JAK_STAT = jak_stat_score_mc
plot_df$Sample = obs_mc_df$batch

H_mc = plot_df$Entropy

```

```{r}


plot_df %>% ggplot(aes(x = JAK_STAT, y = Entropy)) + geom_point(aes(color = Sample)) + geom_smooth(method=lm, color='black') + theme_bw()+
  xlab('JAK/STAT Signaling Score') + ylab('Basal-Luminal Mixing (Entropy)') +  
  xlim(c(-2,2)) + ylim(c(-0, 1.5)) +  
  theme(axis.title.x = element_text(size=20),
        axis.title.y = element_text(size=20),panel.grid.minor = element_blank(),panel.grid.major = element_blank(),
        axis.text.x = element_text(size=20),axis.text.y = element_text(size=20), legend.text = element_text(size=20), legend.title = element_text(size=20)) +
  annotate("text", x=1.38, y=1.45, label = paste("r =", round(cor(plot_df$JAK_STAT, plot_df$Entropy), 2)), size=6)


```


### FIGURE S5D
```{r}

gmt_dir = '/Users/chanj3/Documents_NotToICloud/MSKCC/Research/Prostate_Organoid.LP.publication.010122/ref/'
gmt_fn = paste0(gmt_dir, 'curated.010122.gmt')

gset = gage::readList(gmt_fn)

```


```{r}

mat2 = scale(mat_mc)

rho = c()
for (pathway in names(gset)) {
  if (pathway %in% rownames(gsea_df)) {
    g = str_split(gsea_df[pathway, 'leadingEdge'],',')[[1]]
  } else {
    g=  gset[[pathway]]
  }
  
  if (length(g) > 1) {
    score =  rowMeans(mat2[,g[g %in% colnames(mat2)]])
    rho = c(rho, cor(H_mc, score))
  } else 
    rho = c(rho, NA)
}

names(rho) = names(gset)
rho = rho[!is.na(rho)]
rho = sort(rho)

thr= quantile(rho, .9)

```



```{r}

names(rho[rho > thr])

pathways_interest = c("ABBUD_LIF_SIGNALING_1_UP",
                      "FGFR_signature.Acevedo_etal",
                      "HALLMARK_ANDROGEN_RESPONSE",
                      "HALLMARK_APOPTOSIS",
                      "HALLMARK_INTERFERON_GAMMA_RESPONSE",
                      "HALLMARK_KRAS_SIGNALING_UP",
                      "KEGG_JAK_STAT_SIGNALING_PATHWAY",
                      "PID_SMAD2_3NUCLEAR_PATHWAY",
                      "REACTOME_APC_C_CDH1_MEDIATED_DEGRADATION_OF_CDC20_AND_OTHER_APC_C_CDH1_TARGETED_PROTEINS_IN_LATE_MITOSIS_EARLY_G1",
                      "REACTOME_INTERFERON_ALPHA_BETA_SIGNALING",
                      "WNT_UP.V1_DN")                                                                                                     
                                   
```


```{r, fig.width=2, fig.height=3}
library(ggplot2)
library(scales)

plot_df = data.frame(Correlation = rho, Pathway = names(rho)) %>% dplyr::filter(Pathway %in% pathways_interest) %>% dplyr::filter(Correlation > thr)

plot_df$Pathway[plot_df$Pathway=="REACTOME_APC_C_CDH1_MEDIATED_DEGRADATION_OF_CDC20_AND_OTHER_APC_C_CDH1_TARGETED_PROTEINS_IN_LATE_MITOSIS_EARLY_G1"] = "REACTOME_APC_C_CDH1_MEDIATED_DEGRADATION"

plot_df %>% ggplot(aes(x=reorder(Pathway, Correlation), y = Correlation)) + geom_bar(stat = 'identity', fill='slateblue') + scale_y_continuous(limits=c(0.7,0.825),oob = rescale_none) + theme_bw()  +  
  theme(axis.title = element_text(size=10),panel.grid.minor = element_blank(),panel.grid.major = element_blank(),
        axis.text = element_text(size=10), legend.text = element_text(size=12), axis.text.x = element_text(angle = 90,vjust = 0.5, hjust=1)) + 
  ggtitle('Top 10% Correlated Pathways') +
  ylab('Correlation to Plasticity') + xlab('') #+ geom_vline(xintercept = thr, linetype = 'dashed', color = 'red')


```
````
