```python
fig_dir = '/home/chanj3/data/SCPC_transformation.021120/out.individual_pooled.021120/figures/'
```

```python
for i in samples2:
    adata = adata_dict[i]
    
    norm_unfiltered_df = res_df.loc[adata.obs_names, adata.var_names]
    bad_genes = norm_unfiltered_df.columns.str.contains(
        "^MT-|^MTMR|^MTND|NEAT1|TMSB4X|TMSB10|^RPS|^RPL|^MRP|^FAU$|UBA52|MALAT1")
    norm_df = norm_unfiltered_df.loc[:,~bad_genes]

    n_components=500
    pca = PCA(n_components=n_components, svd_solver='randomized')
    pca.fit(norm_df)

    kneedle = KneeLocator(list(range(len(pca.explained_variance_ratio_))), np.cumsum(pca.explained_variance_ratio_),
                      S=3.0, curve='concave', direction='increasing')
    knee_point = round(kneedle.knee, 3)

    #By Kneepoint
    num_components = 0
    num_components = max(num_components,knee_point)
    
    var_explained = np.cumsum(pca.explained_variance_ratio_)[num_components]
    print('#PCs %d, variance explained %f' % (num_components, var_explained))

    plt.figure(figsize=(5,5))
    plt.plot(np.cumsum(pca.explained_variance_ratio_))
    plt.xlabel('PCs')
    plt.ylabel('% variance explained')
    plt.axvline(knee_point, linestyle='--', color='k')
    plt.ylim(pca.explained_variance_ratio_[0],np.cumsum(pca.explained_variance_ratio_)[-1])
    plt.title('Single cells from %s' % i, fontsize=18)
    plt.savefig(fig_dir + 'PCA.var_explained.%s.single_cell.010122.pdf' % i)
```

```python
fig, axes = plt.subplots(2,3,figsize = (12,8))
for i, ax in zip(samples2, np.ravel(axes)):
    adata = adata_dict[i]
    
    norm_unfiltered_df = res_df.loc[adata.obs_names, adata.var_names]
    bad_genes = norm_unfiltered_df.columns.str.contains(
        "^MT-|^MTMR|^MTND|NEAT1|TMSB4X|TMSB10|^RPS|^RPL|^MRP|^FAU$|UBA52|MALAT1")
    norm_df = norm_unfiltered_df.loc[:,~bad_genes]

    n_components=500
    pca = PCA(n_components=n_components, svd_solver='randomized')
    pca.fit(norm_df)

    kneedle = KneeLocator(list(range(len(pca.explained_variance_ratio_))), np.cumsum(pca.explained_variance_ratio_),
                      S=3.0, curve='concave', direction='increasing')
    knee_point = round(kneedle.knee, 3)

    #By Kneepoint
    num_components = 0
    num_components = max(num_components,knee_point)
    
    var_explained = np.cumsum(pca.explained_variance_ratio_)[num_components]
    print('#PCs %d, variance explained %f' % (num_components, var_explained))

    ax.plot(np.cumsum(pca.explained_variance_ratio_))
    ax.set_xlabel('PCs')
    ax.set_ylabel('% variance explained')
    ax.axvline(knee_point, linestyle='--', color='k')
    ax.set_xlim([-10,210])
    ax.set_ylim(pca.explained_variance_ratio_[0],np.cumsum(pca.explained_variance_ratio_)[210])
    ax.set_title('Single cells from %s' % re.sub('_',' ',i), fontsize=18)

plt.tight_layout()
plt.savefig(fig_dir + 'PCA.var_explained.individual.single_cell.010122.pdf')
```

```python
for i,j in zip(samples, samples2):
    mc_dir = '/home/chanj3/data/SCPC_transformation.metacells.120120/out.cell_line.individual.120120/%s/metacell.final/' % i
    assgn = mmread(mc_dir + 'metacell.assignments.mtx')
    with open(mc_dir + 'metacell.centers.txt', 'r') as f:
        centers = [int(np.float16(i)) for i in f.readlines()]
    
    assgn = np.ravel(np.argmax(assgn, axis=1))
    adata_dict[j].obs.loc[:,'metacell_assignment'] = assgn
    adata_dict[j].obs.loc[:,'metacell_centers'] = [x in centers for x in np.arange(adata_dict[j].shape[0])]
```
