### Figure S4E

```python
labels = ['pval_mc_Basal_Org_1', 'pval_mc_Basal_Org_2', 'pval_mc_Basal_Org_3',
       'pval_mc_Luminal_Org_1', 'pval_mc_Luminal_Org_2']
```

```python
plot_df = adata.obs.loc[adata.obs.batch != 'DHT 0', labels].melt(var_name = 'Cell Type', value_name = 'p-value')
plot_df.loc[:,'Cell Type'] = plot_df.loc[:,'Cell Type'].str.replace('pval_mc_','')
```

```python
sc.set_figure_params(fontsize = 20)
```

```python
sns.set_style('ticks')
```

```python
fig, ax = plt.subplots(1,1, figsize = (5, 5))
sns.boxplot(data = plot_df, x = 'Cell Type', y = 'p-value', ax=ax, 
            palette = ['pink','red','darkred','blue','lightblue'], showfliers = False)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 30)
ax.set_ylabel('Cell type probability\nin mutant cells')
```
