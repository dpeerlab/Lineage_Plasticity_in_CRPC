### Fig 2E

```r
deg_fn = paste0(deg_dir, 'Table_S1.BULK_RNA-seq_P53_RB1_WT_vs_P53_RB1_KO.txt')
```

```r
deg_df = read.table(deg_fn, header=T, sep ='\t', row.names = 'Transcript.ID')
deg_df = deg_df %>% dplyr::group_by(Ensembl) %>% dplyr::filter(Total.counts==max(Total.counts))
colnames(deg_df) = c('Gene','Counts','P.value','FDR','Fold.change')
```

```r
deg_df$log2FC = log2(abs(deg_df$Fold.change))*sign(deg_df$Fold.change)
```

```r
g_interest = c('Krt8','Krt18','Dpp4','Ar','Nkx3-1','Folh1','Trp63','Krt14','Krt5',
               'Cdh1','Snai1','Snai2','Zeb2','Vim','Fgfr1','Fgf2','Fgf7','Fgf10','Fgf13','Wnt5a','Slit3')
```

```r
deg_df$Bonferroni = pmin(1,deg_df$P.value * nrow(deg_df))
```

```r
deg_df$delabel = deg_df$Gene
deg_df$delabel[!deg_df$Gene %in% g_interest] = ''
deg_df$delabel[deg_df$FDR > 0.05 | abs(deg_df$log2FC) < 1] = ''
```

```r
deg_df$diffexpressed = 'black'
deg_df$diffexpressed[deg_df$Bonferroni < 0.05 & abs(deg_df$log2FC) > 1] = 'lightgray'
deg_df$diffexpressed[deg_df$delabel!=''] = 'red'
```

```r
library(ggrepel)

ggplot(data=deg_df %>% dplyr::arrange(delabel), aes(x=log2FC, y=-log2(FDR), col=diffexpressed, label=delabel)) +
        geom_point(size=1) + 
        theme_minimal() +
        geom_text_repel(max.overlaps = Inf, size=6) +
        scale_color_manual(values=c("lightgray", "gray", "red")) +
        geom_vline(xintercept=0, col="gray") + 
        geom_hline(yintercept=0, col="gray") + 
        geom_vline(xintercept=c(-1, 1), col="blue") +
        geom_hline(yintercept=-log2(0.001), col="blue") + theme(text = element_text(size=20),
                                                   plot.title = element_text(hjust = 0.5),
                                                   panel.grid.minor = element_blank(),
                                                  panel.grid.major = element_blank(),
                                                              legend.position = "none") 
```

### Fig 2F

```r
gmt_dir = '/home/chanj3/data/Prostate.LP.publication.010122/ref/'
gmt_fn = paste0(gmt_dir, 'curated.010122.gmt')

gset = gage::readList(gmt_fn)
```

```r
RunGSEA_bulk <- function(deg_fn, gset) {
    set.seed(100)
  deg_df = read.table(deg_fn, header=T, sep ='\t', row.names = 'Transcript.ID')
    deg_df %>% dplyr::group_by(Ensembl) %>% dplyr::filter(Total.counts==max(Total.counts))
  deg_df = deg_df %>% dplyr::arrange(Ensembl,-Total.counts)
  deg_df = deg_df[!duplicated(deg_df$Ensembl),]
  rownames(deg_df) = deg_df$Ensembl
  
  min_fdr = sort(unique(deg_df$FDR.step.up..P53.RB.KO.vs..WT.))[2]
  deg_df$FDR.step.up..P53.RB.KO.vs..WT.[deg_df$FDR.step.up..P53.RB.KO.vs..WT.==0] = min_fdr
  rnk = -log2(deg_df$FDR.step.up..P53.RB.KO.vs..WT.) * (2*(deg_df$Fold.change..P53.RB.KO.vs..WT. > 0) - 1)
  # rnk = -log2(deg_df$fdr) * deg_df$coef
  names(rnk) = toupper(rownames(deg_df))
  score = rnk[!is.na(rnk)] 
  
  fgseaRes <- fgsea(pathways = gset, 
              stats = score,
              minSize=10,
              maxSize=500)#,
              #nperm=10000)
  
  fgseaRes$leadingEdge = sapply(fgseaRes$leadingEdge, function(x) str_c(x,collapse=','))
  return(fgseaRes)
  
}
```

```r
fgseaRes = RunGSEA_bulk(deg_fn, gset)
```

```r
pval_cutoff = 0.2

ind1 = fgseaRes$padj < pval_cutoff
ind2 = abs(fgseaRes$NES) > 1
ind1[is.na(ind1)] = FALSE
ind2[is.na(ind2)] = FALSE

plot_df = fgseaRes[ind1 & ind2,]
```

```r
write.table(plot_df, paste0(gsea_dir, 'GSEA.MUT_WT.bulk.filtered.010122.txt'),quote=F,sep='\t', row.names=F)
```
