<h3> Figure S8A </h3>

```python
'''
CELL TYPE SCORING BY CORRELATION
'''
from scipy.stats import pearsonr
from scipy.stats import gaussian_kde
```

```python
def normalize(data):
    means = np.ravel(data.mean(axis=1))
    std = np.ravel(data.std(axis=1))
    ret = ((data.T - means) / std).T
    return ret
```

```python
ref_dir = '/home/chanj3/data/Prostate.LP.publication.010122/out.bulk_basal_luminal.010122/'

# Load data
bulk_dir = ref_dir
bulk_data = pd.read_csv(bulk_dir + 'bulk_rna_seq.basal_luminal.counts.csv',index_col=0)

# Prepare bulk data
sorted_bulk = bulk_data.loc[:, bulk_data.columns[bulk_data.columns.str.contains('CD49f|CD24')]]
sorted_bulk = sorted_bulk.loc[sorted_bulk.index[sorted_bulk.sum(axis=1) > 10 ],:]
sorted_bulk = (sorted_bulk.divide(sorted_bulk.sum()/1e6))
sorted_bulk = np.log2(sorted_bulk + 1)
sorted_bulk = normalize(sorted_bulk)

cell_types = pd.Series(sorted_bulk.columns.str.replace('-pos.*|MP-[23]', ''),
                       index=sorted_bulk.columns)
```

```python
titles = pd.Series(['Basal CD49f', 'Luminal CD24'], index=['CD49f', 'CD24'])

BL_markers = sorted_bulk.groupby(cell_types,axis=1).mean().sort_values(by = 'CD24', axis = 0)
```

```python
# Correlation with single cells
from sklearn.metrics import pairwise_distances
```

```python
ind_dir='/home/chanj3/data/Prostate.LP.publication.010122/out.metacells.individual.010122/Wk0/MAST.phenograph/'
with open(ind_dir + 'wt_deg.filtered.100121.txt','r') as f:
    wt_deg = [i.strip() for i in f.readlines()]
```

```python
mat = pd.DataFrame(adata.X, index=adata.obs.index, columns = adata.var_names)
mat = mat.loc[:,mat.sum(axis=0)!=0]
mat = mat.loc[:,mat.columns.isin(wt_deg)]

# Single cell data
sc_data = normalize(mat.T ).T

# Common genes
common_genes = sc_data.columns.intersection(sorted_bulk.index)

# Distances
dists = pd.DataFrame(pairwise_distances(sc_data.loc[:, common_genes],
                                        sorted_bulk.loc[common_genes,:].T, metric='correlation'),
                     index=sc_data.index, columns=sorted_bulk.columns)
ct_dists_rls = (1 - dists).groupby(cell_types, axis=1).mean()
```

```python
ct_dists_rls.loc[:,'batch'] = adata.obs.batch
```

```python
sc.set_figure_params(fontsize=10)
sns.set_style('ticks')

fig, ax = plt.subplots(1,1,figsize=(1.5,1.5))
sns.scatterplot(data = ct_dists_rls, x='CD24',y='CD49f', color = 'darkred', edgecolor='white', 
                    s=10, linewidth=.5, ax=ax)
ax.set_xlabel('CD24+ Luminal')
ax.set_ylabel('CD49f+ Basal')
```

### Figure S9D

```python
from sklearn.metrics import pairwise_distances
```

```python
sc.set_figure_params(fontsize=16)
sns.set_style('ticks')
```

```python
dd = pd.DataFrame(0, index=np.arange(1000), columns = adata.obs.batch.cat.categories)
```

```python
n_eigs=6
use_eigs = list(range(1, n_eigs))
eig_vals = np.ravel(adata.uns['diffmap_evals'][use_eigs])
ms_data = pd.DataFrame(adata.obsm['X_diffmap'][:,use_eigs] * (eig_vals / (1-eig_vals)))
ms_data.index = adata.obs_names

cells_b = adata.obs.index[adata.obs.loc[:,'SBM_cluster']==1]
cells_l = adata.obs.index[adata.obs.loc[:,'SBM_cluster']==2]
for sample in adata.obs.batch.cat.categories:
    # Multiscale space
    cells_sample = adata.obs.index[adata.obs.batch == sample]

    b_data = ms_data.loc[cells_sample.intersection(cells_b),:]
    l_data = ms_data.loc[cells_sample.intersection(cells_l),:]

    for n in range(1000):
        sub_mat = pairwise_distances(b_data.iloc[np.random.randint(0, b_data.shape[0], 50),:], 
                                     l_data.iloc[np.random.randint(0, l_data.shape[0], 50),:], 
                                     metric='euclidean')
        dd.loc[n,sample] = np.log10(np.mean(list(np.ravel(sub_mat))))

plot_df = dd.melt(var_name = 'Sample', value_name = 'Log Mean\nDiffusion Distance\nBetween Basal and Luminal')

plot_df.columns = ['Sample','Diffusion_Distance']

fig, ax = plt.subplots(1, 1, figsize = (5,5))

sns.boxplot(data=plot_df, x='Sample', y ='Diffusion_Distance', ax=ax,
           palette = ['gray'] + list(adata.uns['batch_plus_wt_cells_colors'][-5:]))

ax.set_ylabel('Log Mean\nDiffusion Distance', fontsize=25)
ax.set_xlabel('Sample', fontsize=25)

ax.set_xticklabels(adata.obs.batch.cat.categories, rotation = 45, fontsize=25)

ax.tick_params(axis='both', labelsize=20)

ax.set_title('Basal-Luminal Plasticity', fontsize=25)
```
